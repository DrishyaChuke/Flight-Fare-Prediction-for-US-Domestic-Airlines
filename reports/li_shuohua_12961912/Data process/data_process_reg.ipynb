```python
import os
import pandas as pd
import numpy
```

```python
df = pd.read_csv('../../data/processed/combined data/combined.csv')
```

```python
df
```

```python
df.startingAirport.value_counts()
```

```python
# Define the fraction you want to sample
fraction = 0.05

# Sample 25% of each starting airport
df_sampled = df.groupby('startingAirport').apply(lambda x: x.sample(frac=fraction)).reset_index(drop=True)

# Display the sampled DataFrame
df_sampled
```

```python
output_path = '../../data/processed/df_reg.csv'
df_sampled.to_csv(output_path, index=False)

print("Data saved to:", output_path)
```
