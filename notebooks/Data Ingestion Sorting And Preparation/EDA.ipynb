```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'flightDate' is in datetime format
df_cleaned['flightDate'] = pd.to_datetime(df_cleaned['flightDate'], errors='coerce')

# Extract month after confirming 'flightDate' is datetime
df_cleaned['flightMonth'] = df_cleaned['flightDate'].dt.month

# Define high-demand periods
def categorize_seasonal_period(month):
    if month in [6, 7, 8]:
        return 'Summer Vacation'
    elif month == 11:
        return 'Thanksgiving'
    elif month in [12, 1]:
        return 'Christmas/New Year'
    elif month in [3, 4]:
        return 'Spring Break'
    else:
        return 'Other'

df_cleaned['seasonalPeriod'] = df_cleaned['flightMonth'].apply(categorize_seasonal_period)

# Calculate average fare per season
seasonal_fare = df_cleaned.groupby('seasonalPeriod')['totalFare'].mean().reindex(
    ['Summer Vacation', 'Thanksgiving', 'Christmas/New Year', 'Spring Break', 'Other']
)

# Plotting
plt.figure(figsize=(10, 6))
seasonal_fare.plot(kind='bar', color='skyblue')
plt.title('Average Flight Fare During Seasonal High-Demand Periods')
plt.xlabel('Seasonal Period')
plt.ylabel('Average Flight Fare (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

In the United States, flight fares tend to peak during specific times of the year due to seasonal demand, holidays, and vacation schedules. Here’s a breakdown of the most expensive travel periods:

1. Summer Vacation (Late May through Early September)
Peak Period: Memorial Day weekend through Labor Day.
Reason: Many families travel during summer due to school vacations, leading to high demand for both domestic and international flights. July, in particular, often sees high prices.
2. Thanksgiving Week (Late November)
Peak Period: The week of Thanksgiving, with Wednesday before Thanksgiving and the Sunday after being the most expensive.
Reason: Thanksgiving is one of the busiest travel holidays in the U.S., as many people travel to see family. Limited flight availability during this period pushes prices up.
3. Christmas and New Year’s Holiday Season (Mid-December through Early January)
Peak Period: Mid-December through the first week of January.
Reason: The holiday season sees a high volume of travelers visiting family, along with winter vacationers. The demand remains high through New Year’s, making this one of the most expensive times.
4. Spring Break (March and Early April)
Peak Period: Varies by region, but generally from early March to mid-April.
Reason: College and school spring breaks lead to increased demand, especially for flights to popular destinations like Florida, California, and beach resort areas.
5. Holiday Weekends
Examples: Presidents’ Day, Fourth of July, Labor Day, and occasionally MLK Day.
Reason: Long weekends encourage short vacations, increasing demand for flights.
Additional Notes:
Business Travel Peaks: September to November and January to March are peak business travel periods, which can drive prices up for flights during the workweek, especially on popular business routes.
