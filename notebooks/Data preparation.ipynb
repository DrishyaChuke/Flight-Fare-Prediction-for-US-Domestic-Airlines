```python
import zipfile
import os
import pandas as pd
import numpy
```

```python
zip_dir = '/Users/vega7unk/Documents/3rd_Sem_DSI/Adv_ML/AT3/at_3/adv_mla_at3/data/external/ATL'

# Directory to extract CSV files
extract_dir = '/Users/vega7unk/Documents/3rd_Sem_DSI/Adv_ML/AT3/at_3/adv_mla_at3/data/raw/ATL'

# The output CSV file where all data will be merged
output_csv = '/Users/vega7unk/Documents/3rd_Sem_DSI/Adv_ML/AT3/at_3/adv_mla_at3/data/processed/ATL_merged.csv'

os.makedirs(extract_dir, exist_ok=True)
```

```python
# Flag to check if it's the first file (to write headers only once)
first_file = True

# Loop through all zip files in the zip_dir
for zip_file in os.listdir(zip_dir):
    if zip_file.endswith('.zip'):
        # Full path to the zip file
        zip_file_path = os.path.join(zip_dir, zip_file)
        
        # Open and extract all CSV files
        with zipfile.ZipFile(zip_file_path, 'r') as z:
            z.extractall(extract_dir)
            
            # Loop through the extracted files to process CSVs
            for extracted_file in z.namelist():
                if extracted_file.endswith('.csv'):
                    # Full path to the extracted CSV file
                    csv_file_path = os.path.join(extract_dir, extracted_file)
                    
                    # Read the CSV file in chunks (to handle large files)
                    for chunk in pd.read_csv(csv_file_path, chunksize=10000):
                        # Write to output CSV (append mode) without writing headers for subsequent files
                        chunk.to_csv(output_csv, mode='a', header=first_file, index=False)
                        
                        # After the first file is written, set flag to False
                        first_file = False

print(f"All files merged into {output_csv}.")
```
