```python
import os
import pandas as pd
import glob
```

```python
all_airport_files = glob.glob("../../data/raw/itineraries_csv/*.csv")
```

```python
# Define the base directory path
base_path = "../../data/raw/itineraries_csv"

# Create an empty list to store the DataFrames
all_dataframes = []

# Loop through each subfolder in the base directory
for folder_name in os.listdir(base_path):
    subfolder_path = os.path.join(base_path, folder_name)
    
    # Check if the path is a directory (subfolder)
    if os.path.isdir(subfolder_path):
        # Loop through each file in the subfolder
        for file_name in os.listdir(subfolder_path):
            file_path = os.path.join(subfolder_path, file_name)
            
            # Check if the file is a CSV
            if file_name.endswith('.csv'):
                try:
                    # Read the CSV file and append it to the list of DataFrames
                    df = pd.read_csv(file_path)
                    all_dataframes.append(df)
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")

# Combine all DataFrames into one
if all_dataframes:
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    print("Combined all CSV files into a single DataFrame.")
else:
    print("No CSV files found or all failed to load.")
```
