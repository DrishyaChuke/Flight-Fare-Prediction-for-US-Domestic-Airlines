```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
from numpy import mean
from numpy import std
```

```python
X = df.drop(['totalFare'], axis=1)
y = df[['totalFare']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
classifier = AdaBoostRegressor()
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
```

```python
cv = RepeatedKFold(n_splits = 10, n_repeats = 3, random_state = 1)
n_scores = cross_val_score(classifier, X, y, scoring =
    'neg_mean_absolute_error', cv = cv, n_jobs = -1, error_score =
    'raise')
```

```python
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R^2 Score: {r2}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#param_grid = {
#    'n_estimators': [100, 300, 500],                
#    'learning_rate': [0.01, 0.05, 0.1]}
param_grid = {
    'n_estimators': [5, 10, 25, 50, 100, 300, 500],                
    'learning_rate': [0.001, 0.0025, 0.005, 0.01, 0.025, 0.05, 0.1]}
```

```python
random_search = RandomizedSearchCV(
    estimator=classifier,
    param_distributions=param_grid,
    n_iter=100,  # Number of parameter settings sampled
    scoring='neg_mean_squared_error',  # Optimization metric
    cv=5,  # 5-fold cross-validation
    verbose=1,
    n_jobs=-1,  # Use all available cores
    random_state=42
)
```

```python
random_search.fit(X_train, y_train)
```

```python
best_adab = random_search.best_estimator_
print("Best Parameters:", random_search.best_params_)
```

```python
# Make predictions on the test set using the best model
y_pred = best_adab.predict(X_test)
```

```python
# Evaluate the tuned model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R^2 Score: {r2}")
```
